```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_data_list = []

for label in range(10):
  img = cv2.imread(f'/content/drive/MyDrive/ann/handwrite/{label}.png')

  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  #print('Size info :', img_gray.shape)
  #print('Info type :', type(img_gray.shape[0]))
  #print('Tupe:', type(img_gray[0]))
  #plt.subplot(131)
  #plt.imshow(img_gray, cmap='Grays', interpolation='None')
  #print(img_gray)
  img_resize = cv2.resize(img_gray, (28,28))
  #print('Size info :', img_resize.shape)
  #print('Info type :', type(img.shape[0]))
  #plt.subplot(132)
  #plt.imshow(img_resize, cmap='Grays', interpolation='None')
  #print(img_resize)
  img_data = 255 - img_resize.reshape(784)
  plt.figure()
  plt.imshow(img_data.reshape(28,28), cmap='Grays', interpolation='None')
  plt.title(f'label = {label}')
  plt.axis('off')
  plt.show()
  #print(img_data)
  #cv2 waitkey(0)
  img_data_list.append(img_data)

print(img_data_list[9])
```

```python
import csv

f = open('/content/drive/MyDrive/ann/handwrite/handwrite.csv','w', newline='')
writer = csv.writer(f)

for label in range(10):
    img_str = img_data_list[label].astype(str)
    label_img = np.append(label, img_str)
    writer.writerow(label_img)

f.close()

print(label_img)
```
